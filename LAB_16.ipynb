```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style('darkgrid')
fig, ax =plt.subplots(nrows=1,ncols=1)
df=sns.load_dataset('tips')
sns.barplot(x = 'time', y ='total_bill',data = df,palette ='plasma',estimator = np.std,ax=ax).set_title('Bar Plot')
```

```python
sns.set_style('darkgrid')
fig, ax =plt.subplots(nrows=1,ncols=1)
df=sns.load_dataset('tips')
sns.boxplot(x='day',y='total_bill',data=df,hue='smoker',ax=ax).set_title('Box Plot')
sns.violinplot(x='day',y='total_bill',data=df,hue='sex',split = True,ax=ax).set_title('Violin Plot')
sns.stripplot(x='day',y='total_bill',data=df,hue='smoker',jitter = True,dodge = True,ax=ax).set_title('Strip Plot')
```
